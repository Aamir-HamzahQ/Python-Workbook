```python
print("This is a distance converter.")
print("Press 1 to convert from km to miles.")
print("Press 2 to convert from miles to km.")
choice = int(input("Enter your choice:"))

if choice == 1:
  miles = float(input("Enter distance in miles: "))
  km = miles * 1.60934
  print (f"The distance in kilometres is {km:.2f} km")
  #:.2f is used for 2 decimal place


if choice == 2:
  km = float(input("Enter distance in kilometres: "))
  miles = km / 1.60934
  print (f"The distance in miles is {miles:.2f} mi.")
```
